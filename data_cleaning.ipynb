# Data cleaning

### Drop redundant columns

```python
df = df.drop(['href', 'ADAPTABILITY', 'ALL AROUND FRIENDLINESS', 'HEALTH AND GROOMING NEEDS',
              'TRAINABILITY', 'PHYSICAL NEEDS'], axis=1)
```

### Change some columns dtype

```python
columns_to_change = ["Drooling Potential", "Potential For Weight Gain", "Potential For Mouthiness",
                     "Prey Drive", "Tendency To Bark Or Howl", "Tendency To Bark Or Howl"]
df = df.astype({column: pd.Int64Dtype() for column in columns_to_change})
```

# Split and clean columns that have to be splited

```python
re_pattern = "(\d+\.\d+|\d+)" # find ints and floats
```

## Life Span

```python
def split_life_span_to_min_max_life_expectancy(life_span_string):
    
    life_span_string = life_span_string.split('. ')[0]
    
    if re.search(r'Up to', life_span_string):
        min_life_expectancy = None
        max_life_expectancy = int(re.findall(re_pattern, life_span_string)[0])
    elif re.search(r'Starts', life_span_string):
        min_life_expectancy = int(re.findall(re_pattern, life_span_string)[0])
        max_life_expectancy = None
    else:
        split_char = '-' if '-' in life_span_string else 'to'
        
        min_string, max_string = life_span_string.split(split_char)
        
        min_life_expectancy = int(re.findall(re_pattern, min_string)[0])
        max_life_expectancy = int(re.findall(re_pattern, max_string)[0])
    
    return min_life_expectancy, max_life_expectancy
```

```python
df['Min Life Expectancy'], df['Max Life Expectancy'] = zip(*df['Life Span'].map(split_life_span_to_min_max_life_expectancy))
```

```python
df[['Life Span', 'Min Life Expectancy', 'Max Life Expectancy']]
```

```python
df = df.drop('Life Span', axis=1)
```

## Height

```python
df['Height'] = df['Height'].map(lambda x: x if x == "nan" or isinstance(x, float) else x.replace('½','.5'))
```

```python
def split_height_to_min_max_height(height_string):
    split_char = 'to'
    
    def transform_to_height_in_inches(string):
        nums_in_string = re.findall(re_pattern, string)
        
        if len(nums_in_string) == 2: # found values for feet and inches
            return float(nums_in_string[0])*12 + float(nums_in_string[1])
        else:
            if re.search(r'(foot|feet)', string):
                return float(nums_in_string[0]) * 12
            else:
                return float(nums_in_string[0])
        
    if type(height_string) == float or height_string == "nan": ## check if nan
        return None, None    
    
    if re.search(r'(males|females)', height_string):
        height_string = re.sub(r'(males|females)', '', height_string)
        male_height_string, female_height_string = height_string.split('and')
        
        male_min_height, male_max_height = split_height_to_min_max_height(male_height_string)
        female_min_height, female_max_height = split_height_to_min_max_height(female_height_string)
        
        max_height = (male_max_height + female_max_height)/2
        min_height = (male_min_height + female_min_height)/2
        
    elif re.search(r'Up to', height_string):
        min_height = None
        max_height = transform_to_height_in_inches(height_string)
        
    elif re.search(r'From', height_string) and len(height_string.split(split_char)) == 1:
        max_height = None
        min_height = transform_to_height_in_inches(height_string)
    else:
        min_string, max_string = height_string.split(split_char)[:2]
        
        min_height = transform_to_height_in_inches(min_string)
        max_height = transform_to_height_in_inches(max_string)
            
    return min_height, max_height
```

```python
df['Min Height'], df['Max Height'] = zip(*df['Height'].map(split_height_to_min_max_height))
```

```python
df[['Height', 'Min Height', 'Max Height']]
```

```python
df = df.drop('Height', axis=1)
```

## Weight

```python
def split_weight_to_min_max_weight(weight_string):
    
    if type(weight_string) == float: ## check if nan
        return None, None
    
    if re.search(r'Up to', weight_string):
        min_string = None
        max_string = weight_string
    elif re.search(r'(Starts|From)', weight_string):
        max_string = None
        min_string = weight_string
    else:
        split_char = '-' if '-' in weight_string else 'to'
        
        min_string, max_string = weight_string.split(split_char)
    
    strings = [min_string, max_string]
    values = []
    for string in strings:
        if string:
            val = float(re.findall(re_pattern, string)[0])
        else:
            val = None
        values.append(val)
        
    min_weight, max_weight = values
            
    return min_weight, max_weight
```

```python
df['Min Weight'], df['Max Weight'] = zip(*df['Weight'].map(split_weight_to_min_max_weight))
```

```python
df[['Weight', 'Min Weight', 'Max Weight']]
```

```python
df = df.drop('Weight', axis=1)
```
